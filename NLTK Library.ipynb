## Tokenization

```python
mytext =""" Hello Mr. Gosavi, how are you? I hope everything is going well. have a good day, see you soon. """
```

```python
import nltk
nltk.download('punkt')
```

```python
# TOKENIZE SENTENCE

from nltk.tokenize import sent_tokenize
```

```python
print(sent_tokenize(mytext))
```

```python
from nltk.tokenize import word_tokenize
```

```python
mytext = """Python is interesting. Python we're U.S.  $5 million is easy and intersting. \ 
Python is not very difficult."""

tokens=word_tokenize(mytext)
print(tokens)
```

```python
freq = nltk.FreqDist(tokens)

for key , val in freq.items():
    print(str(key)+':'+str(val))
```

```python
#import nltk.classify.util
import nltk
nltk.download('names')
import nltk 
from nltk.classify import NaiveBayesClassifier
from nltk.corpus import names
```

```python
# Load data and training 
names = ([(name, 'male') for name in names.words('male.txt')] + 
    [(name, 'female') for name in names.words('female.txt')])
```

```python
print(names)
```

```python
def gender_features(word): 
    return {'last_letter':word[-1]}
```

```python
train_set = [(gender_features(n), g) for (n,g) in names] 
print(train_set)
```

```python
classifier = nltk.NaiveBayesClassifier.train(train_set) 
```

```python
# Predict
name = input("Enter your name: ")
print(classifier.classify(gender_features(name)))
```
